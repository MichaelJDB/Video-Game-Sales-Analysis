# **Visualisations:**

**Bar Chart - Genres**

```python
import pandas as pd
data = pd.read_csv('../Data/vgsales_cleaned.csv')
```

```python
top_genres = data.groupby('Genre')['Global'].sum().sort_values(ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
top_genres.plot(kind='bar', color='blue')
plt.title('Top 5 Video Game Genres by Global Sales')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This bar chat shows which genres of video games sold the most units worldwide with 'Sports' being the most sold genre. There isn't a huge difference between these five genres which indicates that they are all successful in the video game market and can be great for targeting genres in the future of game development. However, if we wanted to be more precise we could examine the reviews per genre to see if people are really enjoying these video game genres. 

```python
# Load the dataset
data = pd.read_csv('../Data/vgsales_cleaned.csv')

# Group by Genre and calculate average review score
avg_reviews = data.groupby('Genre')['Review'].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
avg_reviews.plot(kind='bar', color='green')
plt.title('Average Review Score by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Review Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This bar chart shows the difference between popularity and critics which helps with those that are enjoyed by the consumers rather than those that are mass purchased. This is useful as it provides companies to see their video game genres are being enjoyed and can continue with creating similar games or select features from previous games that make up those genres to carry over to new ones to target their audiences. 

**Treemap - Regions**

```python
import squarify
data = pd.read_csv('../Data/vgsales_cleaned.csv')

#Sum sales by region
region_sales = data[['North America', 'Europe', 'Japan', 'Rest of World']].sum()

#Create the treemap
plt.figure(figsize=(10, 6))
squarify.plot(sizes=region_sales.values, 
              label=region_sales.index, 
              value=region_sales.values,
              color=['skyblue', 'lightgreen', 'salmon', 'gold'],
              alpha=0.8)
plt.title('Treemap of Video Game Sales by Region')
plt.axis('off')
plt.show()
```

This treemap shows the videos game sales by region which shows the global market sales are distributed across the four major regions. North America dominates the global market with nearly double the sales of Europe. This is useful information as it can be used for regioanl strategies and to show companies the markets they need to focus and with the addition of other factors in the dataset they can improve their goals of continuing or improving their market sales. 

```python
import seaborn as sns

#Load the dataset
data = pd.read_csv('../Data/vgsales_cleaned.csv')

#Create boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Platform', y='Global', data=data)
plt.title('Global Sales Distribution by Platform')
plt.xlabel('Platform')
plt.ylabel('Global Sales (millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This boxplot compares video game sales across different gaming platforms, the median shows the middle sales value for each platform, the box edges are the interquartile range, the whiskers are the typical range of sales and the dots outside the whiskers are outliers which are games that sold either much more or less than the average. 

Wii has the most outliers, close to 80 million which could be diregarded unless its due to a specific game title. This logic can also be applied to other platforms as it shows a high volume of sales for others like PS2, X360, etc but they would need closer examination to see what is causing those high medians. 

THis boxplot is useful as it supports with comparing sales across platlforms, reveals which platforms have potential vs those steady performers and platform tagetting / marketing segmentation. 

```python
#Load dataset
data = pd.read_csv('../Data/vgsales_cleaned.csv')

#Group by Platform and sum global sales
top_platforms = data.groupby('Platform')['Global'].sum().sort_values(ascending=False).head(10)

#plot as horizontal bar chart
plt.figure(figsize=(10, 6))
top_platforms.plot(kind='barh', color='purple')
plt.title('Top 10 Platforms by Total Global Sales')
plt.xlabel('Global Slaes (millions)')
plt.ylabel('Platform')
plt.gca().invert_yaxis() 
plt.tight_layout()
plt.show()

```

This horizontal bar chart supports the previous chart by showing a more precise insight into the video game sales for each platform. The highest sales platform is PS2 followed up by Wii. There is not a huge gap between each of these consoles as they decline in volume.
